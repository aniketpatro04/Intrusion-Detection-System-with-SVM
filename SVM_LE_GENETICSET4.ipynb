```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('kddcup99.csv')
```

```python
dataset.head()
```

```python
#Encoding the 'label' column such that Nomral : 0 and Everything Else is 1. 
#Using the replace method with a replace dictionary.

#Creating a copy of the dataset
df = pd.read_csv('kddcup99.csv')

df['label'].replace({
    'normal' : 0,
    'smurf' : 1,
    'neptune' : 1,
    'back' : 1,
    'satan' : 1,
    'ipsweep' : 1,
    'portsweep' : 1,
    'warezclient' : 1,
    'teardrop' : 1,
    'pod' : 1,
    'nmap' : 1,
    'guess_passwd' : 1,
    'buffer_overflow' : 1,
    'land' : 1,
    'warezmaster' : 1,
    'imap' : 1,
    'rootkit' : 1,
    'loadmodule' : 1,
    'ftp_write' : 1,
    'multihop' : 1,
    'phf' : 1,
    'perl' : 1,
    'spy' : 1
    
}, inplace = True)

df['label'].value_counts()
```

```python
#Encoding the other categorical variables using Label Encoding from Scikit - Learn

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['protocol_type']  = le.fit_transform(df['protocol_type'])

df['service'] = le.fit_transform(df['service'])

df['flag'] = le.fit_transform(df['flag'])

df.head(20)
```

```python
#Creating the Feature Set X and the Target Set Y using the GENETIC SET D Algorithm.


cols = ['src_bytes','dst_bytes','count','diff_srv_rate','dst_host_srv_count','dst_host_same_src_port_rate','label']
df1 = df[cols]
df1
```

## Checking for 10K Data

```python
df2 = df1.head(10000)
```

```python
df2
```

```python
#Creating and the Feature and target Set for 10K Data
x1 = df2.iloc[:, :-1].values
y1 = df2.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train1, X_test1, y_train1, y_test1 = train_test_split(x1, y1, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train1 = sc.fit_transform(X_train1)
X_test1 = sc.transform(X_test1)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(X_train1, y_train1)
```

```python
y_pred1 = classifier.predict(X_test1)
print(np.concatenate((y_pred1.reshape(len(y_pred1),1), y_test1.reshape(len(y_test1),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test1, y_pred1)
print(cm)
accuracy_score(y_test1, y_pred1)
```
